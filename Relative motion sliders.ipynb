```julia
using Reactive, Interact, Plots
```

```julia
@manipulate for xc0 = -5:.05:5, yc0 = -5:.05:5
    x(xc,t) = xc + sin(t)
    y(xc,yc,t) = yc - 1.5*xc*t + 2cos(t)
    tau = -8pi:.1:8pi
    plot(map(t->y(xc0,yc0,t),tau), map(t->x(xc0,t),tau), xlims = (-10,10), ylims = (-7,7), legend = false)
    hline!([0.0])
    plot!([0.0],[0.0],markershape=:star)
end
```
